Exercise 2 : <b>Opening a NetCDF file</b>

```python
import cdms2
DATAPATH = './data/Obs/mo/sst/HadISST/'
# open the nc file throgh cdms2 module
f = cdms2.open(DATAPATH + 'sst_HadISST_Climatology_1961-1990.nc')
# You can query the file
print f.listvariables()
```

```python
# You can “access” the data through file variable
x = f['sst']
# You can get some information about the variables by
x.info()
```

```python
# or read all of it into memory
y = f('sst') # Causion : It will load the whole data into RAM memory 
y.info()
```

```python
  # You can also find out what class the object x or y belong to
print x.__class__
# Close the file
f.close() # do remeber we opened the nc file throgh 'f' file object
```

<center><font face="sans-serif" size=5>Multiple way to retrieve data : <b>Dimension</b></font></center>

* All of it, omitted dimensions are retrieved entirely <br>
<font face="sans-serif" color="Maroon">
data = f('var') </font>

* Specifying dimension type and values <br>
<font face="sans-serif" color="Maroon">
data = f('var', time=(time1,time2)) </font><br>
• Known dimension types : time, level, latitude, longitude (t,z,y,x) 

* Dimension names and values <br>
<font face="sans-serif" color="Maroon">
data = f('var', dimname1=(val1,val2)) </font>

* Sometimes indices are more useful than actual values <br>
<font face="sans-serif" color="Maroon" >
data = f('var', time=slice(index1, index2, step)) </font>

* Relative time

```python
dir(r)
```

* Component time


```python
c = cdtime.comptime(2011, 5, 20)
c
```

```python
dir(c)
```

* You can interchange between component and relative time

```python
c1 = c.torel("days since 2011-1-1")
print "Comptime - ", c
print "Relative - ", c1
```

```python
r1 = r.tocomp()
print "Relative - ", r
print "Component - ", r1
```
